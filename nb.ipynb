```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn

from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
train = pd.read_csv('train.csv')
train = train.drop(columns=["pred edge", "counterparty (naive)", "sm sentiment (naive)", "pnl"]).set_index("trade id")
test = pd.read_csv('test.csv').set_index("trade id").drop("pred edge", axis=1)
```

```python
train = train.join(pd.get_dummies(train[["stock", "location", "dir", "counterparty"]]))
train = train.drop(columns=["stock", "location", "dir", "counterparty"])
train
```

```python
ss = StandardScaler()
x_train = train.drop("realized edge", axis=1)
x_train = ss.fit_transform(x_train)

y_train = train["realized edge"]

pd.DataFrame(x_train)
```

```python
test = test.join(pd.get_dummies(test[["stock", "location", "dir", "counterparty (naive)"]]))
test = test.drop(columns=["stock", "location", "dir", "counterparty (naive)"])
test

x_test = ss.transform(test)
pd.DataFrame(x_test)
test
```

```python
train.drop("realized edge", axis=1).head()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = pd.DataFrame(lr.predict(x_test))
```

```python
y_pred.index.name = "trade id"
y_pred.columns = ["realized edge"]
```

```python
y_pred
```

```python
y_pred.to_csv("pred.csv")
```

# LGBM

```python
import lightgbm as lgb
```

```python
x_train.shape
```

```python
lgb_train = lgb.Dataset(x_train[:175000], y_train[:175000])
lgb_eval = lgb.Dataset(x_train[175000:], y_train[175000:], reference=lgb_train)

# specify your configurations as a dict
params = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': {"rmse"},
    'num_leaves': 1000,
    'learning_rate': .01,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0
}

gbm = lgb.train(params,
                lgb_train,
                num_boost_round=5000,
                valid_sets=lgb_eval,
                callbacks=[lgb.early_stopping(stopping_rounds=100, verbose=True)])
```

```python
lgb_train = lgb.Dataset(x_train, y_train)
lgb_eval = lgb.Dataset(x_train, y_train, reference=lgb_train)

# specify your configurations as a dict
params = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': {"rmse"},
    'num_leaves': 1000,
    'learning_rate': 1,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0
}

gbm = lgb.train(params,
                lgb_train,
                num_boost_round=500,
                valid_sets=lgb_eval,
                callbacks=[lgb.early_stopping(stopping_rounds=50, verbose=True)]
                )
```

```python
pred_lgb = pd.DataFrame(gbm.predict(x_test))
pred_lgb.index.name = "trade id"
pred_lgb.columns = ["realized edge"]
pred_lgb
```

```python
pred_lgb.to_csv("pred_lgb.csv")
```
